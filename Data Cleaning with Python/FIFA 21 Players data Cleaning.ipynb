```python
import re

def convert_height_to_cm(height):
    feet, inches_str = height.split("'")
    feet = int(feet)
    inches = int(re.search(r'\d+', inches_str).group()) 
    height_cm = feet * 30.48 + inches * 2.54
    return height_cm

df['Height_cm'] = df['Height'].apply(convert_height_to_cm)
```

```python
def convert_weight_in_kg(weight):
    numeric_part = re.search(r'\d+', weight)
    
    if numeric_part:
        weight_int = int(numeric_part.group())
        
        weight_kg = weight_int / 2.20462
        return weight_kg
    else:
        return None  
    
df['Weight_kg'] = df['Weight'].apply(convert_weight_in_kg)
```
